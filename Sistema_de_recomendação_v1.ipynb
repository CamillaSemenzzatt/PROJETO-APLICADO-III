```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("projeto_aplicado.db")
cursor = conn.cursor()

# Obter o nome de todas as tabelas no banco de dados
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
tabelas = cursor.fetchall()

# Iterar pelas tabelas e imprimir o nome das colunas
for tabela in tabelas:
    nome_tabela = tabela[0]
    cursor.execute(f"PRAGMA table_info({nome_tabela});")
    colunas = cursor.fetchall()

    print(f"Colunas da tabela {nome_tabela}:")
    for coluna in colunas:
        print(f"  - {coluna[1]}")  # coluna[1] contém o nome da coluna
    print("\n")

conn.close()
```

```python
import sqlite3
import pandas as pd
from IPython.display import display  # Importar a função display

# Nome do banco de dados
db_name = "projeto_aplicado.db"

# Consulta SQL com 3 junções
query_3_juncoes = """
SELECT
    o.municipio,
    o.trimestre,
    o.`Ocupação no trabalho`,
    o.`quantidade de ocupação no trabalho`,
    pc.`Posição na ocupação`,
    pc.`quantidade de posição e categoria`,
    g.Sexo,
    g.`quantidade de gênero`
FROM
    tabela_ocupacao_trabalho_atualizado o
LEFT JOIN
    tabela_posicao_categoria_atualizado pc ON o.municipio = pc.municipio AND o.trimestre = pc.trimestre
LEFT JOIN
    tabela_genero_atualizado g ON o.municipio = g.municipio AND o.trimestre = g.trimestre;
"""

# Consulta SQL com 4 junções
query_4_juncoes = """
SELECT
    e.municipio,
    e.trimestre,
    e.`Nivel de Instrução` AS nivel_instrucao,
    e.`quantidade de escolaridade` AS quantidade_escolaridade,
    p.`posição de ocupação`,
    p.`quantidade de posição de ocupação`,
    c.`Condição de ocupação`,
    c.`quantidade de condição de ocupação`,
    i.Grupo AS grupo_etario,
    i.`quantidade de idade`
FROM
    tabela_escolaridade_atualizado e
LEFT JOIN
    tabela_posicao_atualizado p ON e.municipio = p.municipio AND e.trimestre = p.trimestre
LEFT JOIN
    tabela_condicao_atualizado c ON e.municipio = c.municipio AND e.trimestre = c.trimestre
LEFT JOIN
    tabela_idade_atualizado i ON e.municipio = i.municipio AND e.trimestre = i.trimestre;
"""

# Função para executar a consulta e retornar o DataFrame
def executar_consulta(query, db_name):
    conn = sqlite3.connect(db_name)
    df = pd.read_sql_query(query, conn)
    conn.close()
    return df

# Executar as consultas e armazenar os resultados em DataFrames
df_3_juncoes = executar_consulta(query_3_juncoes, db_name)
df_4_juncoes = executar_consulta(query_4_juncoes, db_name)

# Exibir as tabelas
display(df_3_juncoes)
display(df_4_juncoes)

# Agora você tem os dados das duas tabelas com junção de 3 e junção de 4 em df_3_juncoes e df_4_juncoes, respectivamente.

```

```python
import sqlite3
import pandas as pd
from IPython.display import display

# ... (código anterior - consultas SQL, executar_consulta, carregamento das tabelas) ...

def recomendar_programas_com_porcentagem(municipio, trimestre, df_3_juncoes, df_4_juncoes):
    """Recomenda programas sociais com base nas características do município e trimestre,
       usando duas tabelas separadas."""

    dados_municipio_3 = df_3_juncoes[(df_3_juncoes["municipio"] == municipio) & (df_3_juncoes["trimestre"] == trimestre)]
    dados_municipio_4 = df_4_juncoes[(df_4_juncoes["municipio"] == municipio) & (df_4_juncoes["trimestre"] == trimestre)]

    # ... (restante do código da função) ...

    if dados_municipio_3.empty or dados_municipio_4.empty:
        print(f"Nenhum dado disponível para o município: {municipio}")
        return

    print(f"\nPorcentagem de pessoas que podem se beneficiar dos programas no município: {municipio}")

    total_pessoas = dados_municipio_4["quantidade de idade"].sum()

    if total_pessoas == 0:
        print("Não há registros suficientes para calcular porcentagens.")
        return

    recomendacoes = []

    # Minha Casa Minha Vida
    if "grupo_etario" in dados_municipio_4.columns and dados_municipio_4["grupo_etario"].str.contains("25 a 39 anos").any():
        beneficiarios = dados_municipio_4.loc[dados_municipio_4["grupo_etario"] == "25 a 39 anos", "quantidade de idade"].sum()
        porcentagem = (beneficiarios / total_pessoas) * 100
        recomendacoes.append(f"- Minha Casa Minha Vida: {porcentagem:.2f}% das pessoas estão na faixa etária beneficiada.")

    # Auxílio Emergencial
    if ("grupo_etario" in dados_municipio_4.columns and dados_municipio_4["grupo_etario"].str.contains("18 a 59 anos").any() and
        "Condição de ocupação" in dados_municipio_4.columns and dados_municipio_4["Condição de ocupação"].str.contains("Fora da força de trabalho").any()):
        beneficiarios = dados_municipio_4.loc[dados_municipio_4["grupo_etario"].isin(["18 a 24 anos", "25 a 39 anos", "40 a 59 anos"]), "quantidade de idade"].sum()
        porcentagem = (beneficiarios / total_pessoas) * 100
        recomendacoes.append(f"- Auxílio Emergencial: {porcentagem:.2f}% das pessoas estão fora da força de trabalho e na faixa etária de 18 a 59 anos.")

    # BPC
    if ("grupo_etario" in dados_municipio_4.columns and dados_municipio_4["grupo_etario"].str.contains("60 anos ou mais").any() or
        "Condição de ocupação" in dados_municipio_4.columns and dados_municipio_4["Condição de ocupação"].str.contains("Pessoas com deficiência").any()):
        beneficiarios_idosos = 0
        if "grupo_etario" in dados_municipio_4.columns:  # Verifica se a coluna existe antes de acessar
            beneficiarios_idosos = dados_municipio_4.loc[dados_municipio_4["grupo_etario"] == "60 anos ou mais", "quantidade de idade"].sum()

        beneficiarios_deficiencia = 0
        if "Condição de ocupação" in dados_municipio_4.columns:  # Verifica se a coluna existe antes de acessar
            beneficiarios_deficiencia = dados_municipio_4.loc[dados_municipio_4["Condição de ocupação"].str.contains("Pessoas com deficiência"), "quantidade de condição de ocupação"].sum()

        porcentagem = ((beneficiarios_idosos + beneficiarios_deficiencia) / total_pessoas) * 100
        recomendacoes.append(f"- BPC: {porcentagem:.2f}% das pessoas são idosos ou pessoas com deficiência que podem se beneficiar.")

    # Programas para jovens (18 a 24 anos)
    if "grupo_etario" in dados_municipio_4.columns and dados_municipio_4["grupo_etario"].str.contains("18 a 24 anos").any():
        beneficiarios = dados_municipio_4.loc[dados_municipio_4["grupo_etario"] == "18 a 24 anos", "quantidade de idade"].sum()
        porcentagem = (beneficiarios / total_pessoas) * 100
        recomendacoes.append(f"- Acessuas Trabalho: {porcentagem:.2f}% das pessoas estão na faixa jovem elegível para qualificação.")
        recomendacoes.append(f"- Programa Jovem Aprendiz: {porcentagem:.2f}% das pessoas estão na faixa jovem elegível para qualificação.")
        recomendacoes.append(f"- ID Jovem: {porcentagem:.2f}% das pessoas estão na faixa jovem elegível para qualificação.")
        recomendacoes.append(f"- Pronatec: {porcentagem:.2f}% das pessoas estão na faixa jovem elegível para qualificação.")

    # Brasil Alfabetizado
    if "nivel_instrucao" in dados_municipio_4.columns and dados_municipio_4["nivel_instrucao"].str.contains("Ensino fundamental incompleto ou equivalente").any():
        beneficiarios = dados_municipio_4.loc[dados_municipio_4["nivel_instrucao"] == "Ensino fundamental incompleto ou equivalente", "quantidade_escolaridade"].sum()
        porcentagem = (beneficiarios / total_pessoas) * 100
        recomendacoes.append(f"- Brasil Alfabetizado: {porcentagem:.2f}% das pessoas não completaram o ensino fundamental.")

    # Qualifica Mais
    if "Ocupação no trabalho" in dados_municipio_3.columns and (dados_municipio_3["Ocupação no trabalho"].str.contains("Trabalhadores de serviços, vendedores do comércio em lojas e mercados").any() or
        dados_municipio_3["Ocupação no trabalho"].str.contains("Trabalhadores da produção de bens e serviços industriais").any()):
        beneficiarios = dados_municipio_3.loc[dados_municipio_3["Ocupação no trabalho"].str.contains("Trabalhadores de serviços, vendedores do comércio em lojas e mercados|Trabalhadores da produção de bens e serviços industriais"), "quantidade de ocupação no trabalho"].sum()
        porcentagem = (beneficiarios / total_pessoas) * 100
        recomendacoes.append(f"- Qualifica Mais: {porcentagem:.2f}% das pessoas trabalham em áreas que podem se qualificar.")

    for recomendacao in recomendacoes:
        print(recomendacao)


```

```python
import ipywidgets as widgets
from IPython.display import display

def escolher_municipio_e_trimestre():
    municipios_disponiveis = sorted(df_3_juncoes["municipio"].unique())
    trimestres_disponiveis = sorted(df_3_juncoes["trimestre"].unique())

    # Adiciona um valor vazio no início das opções
    municipios_disponiveis = [''] + municipios_disponiveis
    trimestres_disponiveis = [''] + trimestres_disponiveis

    municipio_widget = widgets.Dropdown(
        options=municipios_disponiveis,
        value='',  # Define o valor inicial como vazio
        description='Município:',
    )

    trimestre_widget = widgets.Dropdown(
        options=trimestres_disponiveis,
        value='',  # Define o valor inicial como vazio
        description='Trimestre:',
    )

    output_widget = widgets.Output()

    def on_change(change):
        with output_widget:
            output_widget.clear_output()  # Limpa a saída anterior
            municipio_escolhido = municipio_widget.value
            trimestre_escolhido = trimestre_widget.value

            if municipio_escolhido and trimestre_escolhido:
                print(f"Município escolhido: {municipio_escolhido}")
                print(f"Trimestre escolhido: {trimestre_escolhido}")
                recomendar_programas_com_porcentagem(municipio_escolhido, trimestre_escolhido, df_3_juncoes, df_4_juncoes)
            else:
                print("Selecione um município e um trimestre.")

    municipio_widget.observe(on_change, names='value')
    trimestre_widget.observe(on_change, names='value')

    display(widgets.VBox([municipio_widget, trimestre_widget, output_widget]))

escolher_municipio_e_trimestre()
```
